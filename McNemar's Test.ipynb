```python
churn1_arr = np.random.choice(churn1_arr, 1000)
churn2_arr = np.random.choice(churn2_arr, 1000)
```

```python
print(len(churn1_arr), len(churn2_arr))
```

```python
#summarize in a table / dataframe

churn_bool_df = pd.DataFrame({'Instance':range(0, len(churn1_arr)),
                                              '1':churn1_arr,
                                              '2':churn2_arr})
churn_bool_df[-5:]
```

```python
pd.set_option('mode.chained_assignment',None)

churn_bool_df['1Yes_2Yes'] = np.zeros(len(churn_bool_df))
churn_bool_df['1Yes_2No'] = np.zeros(len(churn_bool_df))
churn_bool_df['1No_2Yes'] = np.zeros(len(churn_bool_df))
churn_bool_df['1No_2No'] = np.zeros(len(churn_bool_df))

churn_bool_df['1Yes_2Yes'][(churn_bool_df['1']==1) & (churn_bool_df['2']==1)] = 1
churn_bool_df['1Yes_2No'][(churn_bool_df['1']==1) & (churn_bool_df['2']==0)] = 1
churn_bool_df['1No_2Yes'][(churn_bool_df['1']==0) & (churn_bool_df['2']==1)] = 1
churn_bool_df['1No_2No'][(churn_bool_df['1']==0) & (churn_bool_df['2']==0)] = 1
```

```python
YesYes_lst = sum(churn_bool_df['1Yes_2Yes'])
YesNo_lst = sum(churn_bool_df['1Yes_2No'])
NoYes_lst = sum(churn_bool_df['1No_2Yes'])
NoNo_lst = sum(churn_bool_df['1No_2No'])
```

```python
print(YesYes_lst,YesNo_lst,NoYes_lst,NoNo_lst)

table = [[YesYes_lst,YesNo_lst],
         [NoYes_lst, NoNo_lst]]
```

```python
from statsmodels.stats.contingency_tables import mcnemar

table = table

result = mcnemar(table, exact=True)
```

```python
print(result)
```

```python
alpha = 0.05
if result.pvalue > alpha:
    print('Fail to reject H0')
else:
    print('Reject H0 in favor of H1')
```
